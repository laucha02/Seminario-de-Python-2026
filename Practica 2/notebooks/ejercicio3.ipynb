```python
review = """La película sigue a un grupo de astronautas que viajan a Marte
en una misión de rescate. El capitán Torres lidera al equipo a través
de tormentas solares y fallos en el sistema de navegación. Al llegar
a Marte descubren que la base está abandonada y los suministros
destruidos. Torres decide sacrificar la nave nodriza para salvar
al equipo y logran volver a la Tierra en una cápsula de emergencia.
El final revela que Torres sobrevivió gracias a un pasaje secreto."""
```

```python
words = input('Ingrese las palabras spoiler (separadas por coma): ')
words_list = []
for palabra in words.split(","):
    words_list.append(palabra.strip().lower())

review_list = review.split(" ")

resu = []

for palabra in review_list:

    clean_word = palabra.strip(".").lower()
    divido = palabra.split("\n")
    if ((len(divido) == 2) and (divido[0].strip(".").lower() in words_list)):  ##Caso específico para procesar las palabras antes y despues del salto de linea
        censura = "*" * len(divido[0])
        if (divido[0].endswith(".")):
            censura = censura + "."
        censura = censura + "\n" + divido[1]
        
        resu.append(censura)
    elif ((len(divido) == 2) and (divido[1].strip(".") in words_list)):
        censura = (divido[0] + "\n")
        censura = censura + ("*" * len(divido[1]))
        if (divido[1].endswith(".")):
            censura = censura + "."

        resu.append(censura)
    elif (clean_word in words_list):
        censura = "*" * len(clean_word)

        if (palabra.endswith(".")): ##Agrega signos de puntuaciones (lo agrego en caso que se cambie la palabra a censurar)
            censura = censura + "."
    
        resu.append(censura)
    else:
        resu.append(palabra)
    
review_final = " ".join(resu)
print(review_final)
```
