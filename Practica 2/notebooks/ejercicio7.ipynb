```python
import random

names = input('Ingrese los participantes (separados por coma): ')
names_l = names.split(", ")
names_list = [n.lower() for n in names_l]
if (len(names_list) != len(set(names_list))):
    print('Error: hay nombres duplicados.')
elif (len(names_list) < 3):
    print('Error: se necesita al menos 3 participantes.')

pareja = {}
total = len(names_list)

random.shuffle(names_list)

for i in range(total):
    a = names_list[i]
    b = names_list[(i + 1) % total] # Cuando llegue al ultimo, total % total = 0

    pareja[a] = b

print('Sorteo de amigo invisible:')
for elem in pareja:
    print(f'{elem.capitalize()} -> {pareja[elem].capitalize()}')
```
