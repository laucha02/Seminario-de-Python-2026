```python
dic = {
    "local" : [500,1000,2000],
    "regional" : [1000,2500,4500],
    "nacional" : [2000,5000,8000]
}
weight = float(input('Ingrese el peso del paquete (kg): '))
area = input('Ingrese la zona de destino (local/regional/nacional): ')
if (area in dic):
    if (weight < 1):
        costo = dic[area][0]
    elif (1 < weight < 5):
        costo = dic[area][1]
    else:
        costo = dic[area][2]
    
    print(f'Costo de envío: ${costo}')
else:
    print('Zona no válida. Las zonas disponibles son: local, regional, nacional')
```
