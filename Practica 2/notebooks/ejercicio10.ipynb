```python
dic = inicializar_diccionario(rounds)

cont = 0

for round in rounds:
    name_max = ""
    score_max = -1
    cont += 1
    for name,judges in round['scores'].items():
        
        punt_temp = sum(judges.values())

        dic[name]['total_points'] += punt_temp

        if (punt_temp > dic[name]['points_best_round']):
            dic[name]['points_best_round'] = punt_temp

        if punt_temp > score_max:
            score_max = punt_temp
            name_max = name
        
    dic[name_max]['rounds_won'] += 1
        
    
    print(f'Ronda {cont} - {round['theme']}:')
    print(f'Ganador: {name_max} ({score_max} pts)')

    sort = sorted(dic.items(), key=lambda x : x[1]['total_points'], reverse=True)
    imprimir_posiciones(sort)



imprimir_tabla_final(sort, len(rounds))


```
