```python
playlist = [
{"title": "Bohemian Rhapsody", "duration": "5:55"},
{"title": "Hotel California", "duration": "6:30"},
{"title": "Stairway to Heaven", "duration": "8:02"},
{"title": "Imagine", "duration": "3:07"},
{"title": "Smells Like Teen Spirit", "duration": "5:01"},
{"title": "Billie Jean", "duration": "4:54"},
{"title": "Hey Jude", "duration": "7:11"},
{"title": "Like a Rolling Stone", "duration": "6:13"},
]

total_minutes = 0
total_seconds = 0

max_total_tiempo = -1 
min_total_tiempo = 100000

song_max_total_tiempo = ""
song_min_total_tiempo = ""

length_min_total_tiempo = 0
length_max_total_tiempo = 0
```

```python
for song in playlist:

    duracion = song["duration"]
    length_list = duracion.split(':')
    total_minutes +=  int(length_list[0])
    total_seconds +=  int(length_list[1])

    temp_total_tiempo = int(length_list[0]) * 60 + int(length_list[1])
    if (temp_total_tiempo > max_total_tiempo):
        max_total_tiempo = temp_total_tiempo
        song_max_total_tiempo = song["title"]
        length_max_total_tiempo = song["duration"]
    elif (temp_total_tiempo < min_total_tiempo):
        min_total_tiempo = temp_total_tiempo
        song_min_total_tiempo = song["title"]
        length_min_total_tiempo = song["duration"]

minutes_final = (total_seconds // 60) + total_minutes
seconds_final = total_seconds % 60

print(f'Duración total: {minutes_final}m {seconds_final}s')
print(f'Canción más larga: {song_max_total_tiempo} ({length_max_total_tiempo})')
print(f'Canción más corta: {song_min_total_tiempo} ({length_min_total_tiempo})')
```
