```python
phrase = input('Ingrese un mensaje: ')
desp = int(input('Ingrese el desplazamiento: '))
phrase_crypted = ""
for letter in phrase:
    if letter.isalpha():
        asc = ord(letter) + desp
        phrase_crypted += chr(asc)
    else:
        phrase_crypted += letter

print(f'Mensaje cifrado: {phrase_crypted}')
print(f'Mensaje descifrado: {phrase}')

```
