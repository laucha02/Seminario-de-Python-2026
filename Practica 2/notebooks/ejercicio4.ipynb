```python

email = input("Ingrese un mail: ")


if (email.count('@') == 1):
    mail_splitted = email.split('@')
    if (len(mail_splitted[0]) > 1):
        if(mail_splitted[1].count('.') >= 1):
            if not (email.startswith("@") or (email.startswith(".") or (email.endswith("@")) or (email.endswith(".")))):
                mail = email.split(".")
                mail2 = mail[len(mail)-1]
                if (len(mail2) >= 2):
                    print('El email es válido.')
                else:
                    print('El email no es válido (5).')
            else:
                print('El email no es válido (4).')
        else:
            print('El email no es válido (3).')        
    else:
        print('El email no es válido (2).')
else:
    print('El email no es válido (1).')
```
